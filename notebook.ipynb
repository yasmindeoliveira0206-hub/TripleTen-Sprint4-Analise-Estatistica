## Receita

[Da mesma forma que você estudou o comportamento dos usuários, descreva estatisticamente as receitas dos planos.]

```python
if 'minutes_included' not in df_final.columns:
    df_final = df_final.merge(
        df_plans,
        left_on='plan',
        right_on='plan_name',
        how='left'
    )


def custo_chamada(row):
    excesso = row['minutos'] - row['minutes_included']
    return excesso * row['usd_per_minute'] if excesso > 0 else 0

def custo_mensagem(row):
    excesso = row['mensagens'] - row['messages_included']
    return excesso * row['usd_per_message'] if excesso > 0 else 0

def custo_dados(row):
    excesso = row['internet_mb'] - row['mb_per_month_included']
    return (excesso / 1024) * row['usd_per_gb'] if excesso > 0 else 0

df_final['custo_chamada'] = df_final.apply(custo_chamada, axis=1)
df_final['custo_mensagem'] = df_final.apply(custo_mensagem, axis=1)
df_final['custo_dados'] = df_final.apply(custo_dados, axis=1)

df_final['receita'] = (
    df_final['custo_chamada'] +
    df_final['custo_mensagem'] +
    df_final['custo_dados'] +
    df_final['usd_monthly_pay']
)  

receita_mensal = df_final.groupby(['year_month', 'plan'])['receita'].mean().reset_index()
receita_pivot = receita_mensal.pivot(
    index='year_month',
    columns='plan',
    values='receita'
)

ax = receita_pivot.plot(kind='bar', figsize=(10,6))
plt.xticks(rotation=45)
plt.xlabel('Mês')
plt.ylabel('Receita total ($)')
plt.grid(axis='y', alpha=0.4)
plt.show()
```

## Teste hipóteses estatísticas

[Teste a hipótese de que a receita média dos usuários dos planos Ultimate e Surf são diferentes.]

[Formule as hipóteses nula e alternativa, escolha o teste estatístico, escolha o valor alfa.]

```python

#Hipotese nula: Afurma que os planos possiem receita média iguais
#Hipotese Alternativa: Afirma que os planos possuem receita média diferentes

from scipy import stats as st


receita_surf = df_final[df_final['plan'] == 'surf']['receita']
receita_ultimate = df_final[df_final['plan'] == 'ultimate']['receita']
alpha = 0.05


test_var = st.levene(receita_surf,receita_ultimate)
if test_var.pvalue <=alpha: 
    equalvar_result = False
else: 
    equalvar_result = True

t_test = st.ttest_ind(receita_surf, receita_ultimate, equal_var=equalvar_result)

if t_test.pvalue<=alpha:
    print('Rejeitamos a hipótese nula, receita média dos usuários dos planos Ultimate e Surf são diferentes')
else:
    print('Não há informações suficientes para rejeitar a hiposete nula')
    
```
